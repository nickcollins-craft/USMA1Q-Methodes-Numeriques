   `245, 251, 238, 260, 249`  

3. **Taux d'allongement** (%) selon la formule :  
   $$
   A = \frac{L_f - L_0}{L_0} \times 100
   $$  
   avec $L_0 = 50{,}0$ mm et $L_f = 63{,}5$ mm.

4. **Volume d'un cylindre** :  
   $$
   V = \pi r^2 h
   $$  
   avec $r = 12{,}5$ mm et $h = 80$ mm.  
   *(Utilisez `3.14159` pour π, ou ajouter la ligne `import math` au debut du block et utilisez `math.pi` où c'est requi)*

3. **Conversion d'unité de temps** : une expérience dure `3 875` secondes.  
   Affichez la durée au format `H h MM min SS s`.

````python
# Exercice 2.3 — solution

# ── 1. Parité ────────────────────────────────────────────────────────────────
# Un nombre est pair si son reste par division par 2 vaut 0
n = 7
est_pair = (n % 2 == 0)
print("n =", n, "est pair :", est_pair)

n = 12
est_pair = (n % 2 == 0)
print("n =", n, "est pair :", est_pair)

# ── 2. Partitionnement d'éprouvettes ─────────────────────────────────────────
nb_eprouvettes = 53
taille_groupe = 8

nb_groupes_complets = nb_eprouvettes // taille_groupe  # division entière
nb_restantes        = nb_eprouvettes % taille_groupe   # reste

print("Groupes complets :", nb_groupes_complets)
print("Éprouvettes restantes :", nb_restantes)

# ── 3. Conversion de secondes ────────────────────────────────────────────────
duree_s = 3875

heures   = duree_s // 3600         # 3600 s dans une heure
minutes  = (duree_s % 3600) // 60  # secondes restantes divisées par 60
secondes = duree_s % 60            # secondes après extraction des minutes

print("Durée :", heures, "h", minutes, "min", secondes, "s")
````

### Exercice 4.1 — Patron accumulateur

Voici une liste de températures de traitement thermique (en °C) :
```python
temperatures = [820, 835, 810, 850, 825, 840, 815, 830, 845, 820]
```

En utilisant une boucle `for` et le patron accumulateur, calculez :

1. La **somme** de toutes les températures.
2. La **température minimale**.
3. La **température maximale**.
4. Le **nombre de températures** strictement supérieures à 830 °C.

*Rappel du patron accumulateur :*
```python
total = 0
for v in liste:
    total += v
```

````python
# Exercice 4.1 — solution
temperatures = [820, 835, 810, 850, 825, 840, 815, 830, 845, 820]

# ── 1. Somme ─────────────────────────────────────────────────────────────────
total = 0
for t in temperatures:
    total += t  # accumuler
print("Somme :", total, "°C")

# ── 2. Minimum ───────────────────────────────────────────────────────────────
mini = temperatures[0]  # initialisation avec le premier élément
for t in temperatures:
    if t < mini:
        mini = t
print("Minimum :", mini, "°C")

# ── 3. Maximum ───────────────────────────────────────────────────────────────
maxi = temperatures[0]
for t in temperatures:
    if t > maxi:
        maxi = t
print("Maximum :", maxi, "°C")

# ── 4. Comptage > 830 °C ─────────────────────────────────────────────────────
compteur = 0
for t in temperatures:
    if t > 830:
        compteur += 1
print("Valeurs > 830 °C :", compteur)
````

### Exercice 4.2 — Statistiques avec une boucle `while`

Reprenez la liste `temperatures` de l'exercice précédent. Cette fois, **utilisez une boucle `while`** (avec un indice `i`) pour calculer la somme, puis la moyenne.

Cet exercice illustre que `for` et `while` peuvent accomplir la même tâche — l'un ou l'autre est simplement plus naturel selon la situation.

````python
# Exercice 4.2 — solution
temperatures = [820, 835, 810, 850, 825, 840, 815, 830, 845, 820]

total = 0
i = 0

while i < len(temperatures):
    total += temperatures[i]  # accumuler la valeur à l'indice i
    i += 1                    # avancer l'indice (ne pas oublier !)

moyenne = total / len(temperatures)
print("Moyenne :", round(moyenne, 1), "°C")
````

### Exercice 4.3 — Moyenne et écart-type par boucle

Soit les dix mesures de dureté Brinell suivantes :
```python
durete = [245, 251, 238, 260, 249, 253, 247, 255, 241, 258]
```

À l'aide de boucles `for`, calculez **sans utiliser** `sum()` ni les fonctions de `math` ou `statistics` :

1. La **somme** et la **moyenne** $\bar{x}$.
2. La **variance** (population) :
   $$s^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2$$
3. L'**écart-type** $s = \sqrt{s^2}$ *(utilisez `** 0.5` ou `** (1/2)`)*.

Vérifiez vos résultats :
- Moyenne ≈ 249.7 HB
- Écart-type ≈ 6.97 HB

> *Indication :* deux boucles séparées sont nécessaires — la première calcule la moyenne, la seconde calcule la somme des carrés des écarts.

````python
# Exercice 4.3 — solution
durete = [245, 251, 238, 260, 249, 253, 247, 255, 241, 258]
n = len(durete)

# ── 1. Somme et moyenne ──────────────────────────────────────────────────────
total = 0
for v in durete:
    total += v
moyenne = total / n
print("Moyenne      :", round(moyenne, 2), "HB")

# ── 2. Variance (population, diviseur n) ────────────────────────────────────
somme_carres = 0
for v in durete:
    somme_carres += (v - moyenne) ** 2  # carré de l'écart à la moyenne
variance = somme_carres / n
print("Variance     :", round(variance, 4), "HB2")

# ── 3. Écart-type = racine carrée de la variance ─────────────────────────────
ecart_type = variance ** 0.5
print("Écart-type   :", round(ecart_type, 2), "HB")
````

### Exercice 4.4 — Suite de Fibonacci et `while`

La **suite de Fibonacci** est définie par la relation de récurrence :
$$F_0 = 0,\quad F_1 = 1,\quad F_n = F_{n-1} + F_{n-2} \text{ pour } n \geq 2$$

1. Utilisez une boucle `while` pour calculer et afficher tous les termes de Fibonacci **inférieurs à 500**.
2. Comptez combien il y en a.
3. *(Bonus)* Affichez pour chaque terme $F_n$ le rapport $\frac{F_n}{F_{n-1}}$ (sauf pour $F_0$).  
   Que remarquez-vous quand $n$ devient grand ?

````python
# Exercice 4.4 — solution

# ── 1 & 2. Termes de Fibonacci < 500 ────────────────────────────────────────
a = 0  # terme courant
b = 1  # terme suivant
compteur = 0

while a < 500:
    print(a)
    compteur += 1
    a, b = b, a + b  # mise à jour simultanée : nouveau a = ancien b,
                     #                          nouveau b = a + b

print("Nombre de termes < 500 :", compteur)

# ── 3. (Bonus) Rapport F_n / F_{n-1} ────────────────────────────────────────
# Ce rapport tend vers le nombre d'or phi = 1.6180...
print("Rapports F_n / F_(n-1) :")
a, b = 0, 1
while b < 500:
    if a > 0:   # éviter la division par zéro pour F_0
        rapport = b / a
        print("  F =", b, "  rapport =", round(rapport, 6))
    a, b = b, a + b
````

## Partie 5 — Mini-projet guidé : Contrôle de deux lots de pièces forgées

### Scénario

Vous travaillez dans un bureau de contrôle qualité. Un four de forge a produit **deux lots** de 10 pièces chacun (lot A et lot B).
Vous disposez des mesures de dureté Brinell (HB) relevées sur chaque pièce des deux lots.

Le cahier des charges impose :
- Dureté **moyenne** $\geq$ 240 HB
- **Écart-type** $\leq$ 15 HB *(dispersion maîtrisée)*

**Votre mission :**

1. Stocker les deux lots dans une liste de listes.
2. À l'aide de **boucles imbriquées** (boucle extérieure sur les lots, boucle intérieure sur les valeurs), calculer la **moyenne** et l'**écart-type** de chaque lot.
3. Afficher si chaque lot est **conforme** ou **non conforme** aux deux critères.
4. Identifier les pièces **individuellement hors tolérance** (HB < 230 ou HB > 270) dans chaque lot.
5. Rédiger une **analyse d'ingénierie** (cellule Markdown) — les deux lots sont-ils acceptables ?


````python
# Mini-projet — données
# Deux lots de 10 mesures de dureté Brinell (HB)
lot_a = [242, 248, 235, 255, 251, 238, 246, 260, 244, 250]
lot_b = [239, 253, 241, 257, 245, 252, 237, 249, 243, 258]

# On regroupe les deux lots dans une liste de listes
lots = [lot_a, lot_b]

# Seuils du cahier des charges
seuil_moyenne   = 240.0  # HB minimum
seuil_ecart_max = 15.0   # HB maximum

# Limites de tolérance individuelle
tolerance_min = 230  # HB
tolerance_max = 270  # HB
````

````python
# ── Étape 2 : Statistiques — boucles imbriquées ──────────────────────────────
moyennes = []
ecarts = []

for i in range(len(lots)):
    lot = lots[i]
    n = len(lot)

    # Calcul de la moyenne (boucle intérieure)
    total = 0
    for v in lot:
        total += v
    moy = total / n
    moyennes.append(moy)

    # Calcul de la variance puis de l'écart-type (boucle intérieure)
    somme_carres = 0
    for v in lot:
        somme_carres += (v - moy) ** 2
    ecart = (somme_carres / n) ** 0.5
    ecarts.append(ecart)

    print("Lot", i + 1, "— moyenne :", round(moy, 2), "HB  | écart-type :", round(ecart, 2), "HB")
````

````python
# ── Étape 3 : Conformité de chaque lot ───────────────────────────────────────

for i in range(len(lots)):
    conforme_moy   = moyennes[i] >= seuil_moyenne
    conforme_ecart = ecarts[i]   <= seuil_ecart_max

    if conforme_moy and conforme_ecart:
        print("Lot", i + 1, ": Conforme")
    else:
        print("Lot", i + 1, ": Non conforme")
        if not conforme_moy:
            print("  - Moyenne =", round(moyennes[i], 2), "HB < seuil de", seuil_moyenne, "HB")
        if not conforme_ecart:
            print("  - Ecart-type =", round(ecarts[i], 2), "HB > seuil de", seuil_ecart_max, "HB")
````

````python
# ── Étape 4 : Pièces hors tolérance individuelle ─────────────────────────────

for i in range(len(lots)):
    print("Lot", i + 1, "— pièces hors tolérance [", tolerance_min, "-", tolerance_max, "HB] :")
    nb_hors = 0
    for j in range(len(lots[i])):
        v = lots[i][j]
        if v < tolerance_min or v > tolerance_max:
            print("  Pièce", j + 1, ": HB =", v)
            nb_hors += 1
    if nb_hors == 0:
        print("  Aucune pièce hors tolérance.")
````
