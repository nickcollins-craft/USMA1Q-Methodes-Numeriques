### Exercice 1.1 — Fonction simple avec validation

Écrivez une fonction `aire_section(diametre)` qui calcule l'**aire de la section transversale** d'une éprouvette cylindrique (en mm²).

$$A = \pi \left(\frac{d}{2}\right)^2$$

**Exigences :**
1. La fonction doit inclure une **docstring** décrivant les paramètres et la valeur de retour.
2. Si `diametre` est négatif ou nul, la fonction doit lever une `ValueError` avec un message explicite.
3. Testez avec : `diametre = 10.0` (résultat attendu : ≈ 78.54 mm²) et `diametre = -5` (attendu : `ValueError`).

*Indice :* utilisez `import math` et `math.pi`, ou `3.14159265`.

````python
# Exercice 1.1 — solution
import math


def aire_section(diametre):
    """
    Calcule l'aire de la section transversale d'une éprouvette cylindrique.

    Paramètres
    ----------
    diametre : float - Diamètre en mm. Doit être strictement positif.

    Retourne
    --------
    aire : float - Aire en mm².
    """
    if diametre <= 0:
        raise ValueError("Le diamètre doit être strictement positif, reçu : " + str(diametre))
    return math.pi * (diametre / 2) ** 2


# Tests
print(round(aire_section(10.0), 2))   # → 78.54
# print(aire_section(-5))              # → ValueError (décommenter pour tester)
````

### Exercice 1.2 — Fonctions statistiques

Écrivez deux fonctions de statistique en utilisant des boucles `for`.

**`moyenne(valeurs)`**  
Retourne la moyenne arithmétique d'une liste. Lève `ValueError` si la liste est vide.

**`ecart_type(valeurs)`**  
Retourne l'écart-type de la **population** (diviseur $n$) :
$$s = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2}$$
Appelle `moyenne()` en interne. Lève `ValueError` si la liste a moins de 2 éléments.

**Testez avec :**
```python
durete = [245, 251, 238, 260, 249, 253, 247, 255, 241, 258]
```
Résultats attendus : moyenne ≈ 249.7 HB, écart-type ≈ 6.77 HB.

````python
# Exercice 1.2 — solution


def moyenne(valeurs):
    """
    Calcule la moyenne arithmétique d'une liste de nombres.

    Paramètres
    ----------
    valeurs : list - Liste de nombres (int ou float). Ne doit pas être vide.

    Retourne
    --------
    moy : float - Moyenne arithmétique de la liste.
    """
    if len(valeurs) == 0:
        raise ValueError("La liste est vide — la moyenne est indéfinie.")
    total = 0
    for v in valeurs:
        total += v
    return total / len(valeurs)


def ecart_type(valeurs):
    """
    Calcule l'écart-type de la population (diviseur n) d'une liste de nombres.

    Paramètres
    ----------
    valeurs : list - Liste d'au moins 2 nombres (int ou float).

    Retourne
    --------
    s : float - Écart-type de la population (diviseur n).
    """
    if len(valeurs) < 2:
        raise ValueError("La liste doit contenir au moins 2 éléments.")
    moy = moyenne(valeurs)
    somme_carres = 0
    for v in valeurs:
        somme_carres += (v - moy) ** 2
    return (somme_carres / len(valeurs)) ** 0.5


# Tests
durete = [245, 251, 238, 260, 249, 253, 247, 255, 241, 258]
print("Moyenne     : ", round(moyenne(durete), 2), "HB")     # → 249.7
print("Écart-type  : ", round(ecart_type(durete), 2), "HB")  # → 6.77
````

### Exercice 1.3 — Contrôle de conformité

Écrivez une fonction `passe_spec(valeur, seuil_min, seuil_max)` qui retourne `True` si `seuil_min ≤ valeur ≤ seuil_max`, et `False` sinon.

**Exigences supplémentaires :**
1. Lever `ValueError` si `seuil_min > seuil_max` (seuils incohérents).
2. Écrire une fonction **`rapport_lot(valeurs, seuil_min, seuil_max)`** qui appelle `passe_spec()` sur chaque élément d'une liste et affiche :
   - Le nombre de pièces conformes et non conformes
   - Le taux de conformité en pourcentage

**Testez avec les deux lots ci-dessous (seuils : 230–270 HB) :**
```python
lot_a = [242, 248, 235, 255, 251, 238, 246, 260, 244, 250]
lot_b = [225, 271, 240, 265, 232, 275, 248, 258, 229, 252]
```

````python
# Exercice 1.3 — solution


def passe_spec(valeur, seuil_min, seuil_max):
    """
    Vérifie si une valeur est dans l'intervalle [seuil_min, seuil_max].

    Paramètres
    ----------
    valeur    : float - Valeur à tester.
    seuil_min : float - Borne inférieure (incluse).
    seuil_max : float - Borne supérieure (incluse).

    Retourne
    --------
    conforme : bool - True si seuil_min <= valeur <= seuil_max, False sinon.
    """
    if seuil_min > seuil_max:
        raise ValueError("seuil_min doit être <= seuil_max, reçu : " + str(seuil_min) + " > " + str(seuil_max))
    return seuil_min <= valeur <= seuil_max


def rapport_lot(valeurs, seuil_min, seuil_max):
    """
    Affiche un rapport de conformité pour une liste de valeurs.

    Paramètres
    ----------
    valeurs   : list   Liste de mesures.
    seuil_min : float  Borne inférieure de conformité.
    seuil_max : float  Borne supérieure de conformité.

    Retourne
    --------
    None
    """
    nb_conformes = 0
    for v in valeurs:
        if passe_spec(v, seuil_min, seuil_max):
            nb_conformes += 1
    nb_non_conformes = len(valeurs) - nb_conformes
    taux = nb_conformes / len(valeurs) * 100
    print("Conformes     : ", nb_conformes)
    print("Non conformes : ", nb_non_conformes)
    print("Taux          : ", round(taux, 1), "%")


lot_a = [242, 248, 235, 255, 251, 238, 246, 260, 244, 250]
lot_b = [225, 271, 240, 265, 232, 275, 248, 258, 229, 252]

print("Lot A : ")
rapport_lot(lot_a, 230, 270)

print("Lot B : ")
rapport_lot(lot_b, 230, 270)
````

### Exercice 2.1 — Découpage (*slicing*)

Soit la liste suivante de résistances à la traction (MPa) mesurées pour une série d'éprouvettes :

```python
Rm = [412, 438, 425, 460, 399, 445, 418, 451, 407, 432, 461, 395]
```

**Sans boucle**, utilisez le découpage pour :

1. Extraire les **5 premières** mesures.
2. Extraire les **3 dernières** mesures (utiliser un indice négatif).
3. Extraire les mesures d'**indice pair** (0, 2, 4…) — une mesure sur deux à partir de la première.
4. Afficher la liste en **ordre inverse**.
5. Extraire les mesures de l'**indice 3 à l'indice 8** (inclus).

> *Rappel :* `liste[debut:fin]` — `debut` inclus, `fin` **exclus**.

````python
# Exercice 2.1 — solution
Rm = [412, 438, 425, 460, 399, 445, 418, 451, 407, 432, 461, 395]

# 1. Cinq premières mesures — indices 0 à 4 (fin 5 exclue)
print("1. Cinq premières  : ", Rm[:5])

# 2. Trois dernières mesures — indices -3, -2, -1
print("2. Trois dernières : ", Rm[-3:])

# 3. Indices pairs 0, 2, 4… — pas de 2 depuis le début
print("3. Indices pairs   : ", Rm[::2])

# 4. Ordre inverse — pas de -1
print("4. Inversée        : ", Rm[::-1])

# 5. Indices 3 à 8 inclus → slice 3:9 (fin exclue, donc 9 doit être écrit pour inclure l'indice 8)
print("5. Indices 3–8     : ", Rm[3:9])
````

### Exercice 2.2 — Compréhensions de listes

Voici une liste de limites d'élasticité mesurées sur des profilés d'acier S355 (MPa) :

```python
Re = [348, 362, 341, 375, 330, 358, 369, 327, 381, 355, 344, 370, 337, 363, 352]
```

En utilisant des **compréhensions de listes**, répondez aux questions suivantes :

1. Créez une liste `Re_gpa` contenant chaque valeur convertie en GPa (diviser par 1000).
2. Créez une liste `Re_conforms` ne contenant que les valeurs **≥ 355 MPa** (valeur nominale S355).
3. Calculez le **taux de conformité** (%) à partir de ces deux listes.
4. Créez une liste `Re_centre` contenant les **écarts à la moyenne** `(v - moy)` pour chaque valeur.  
   Appelez votre fonction `moyenne()` de la Partie 1 pour obtenir la moyenne.

> *Vérification :* la somme de `Re_centre` doit être très proche de 0.

````python
# Exercice 2.2 — solution
# (Les fonctions moyenne() et ecart_type() de l'exercice 1.2 sont déjà définies.)

Re = [348, 362, 341, 375, 330, 358, 369, 327, 381, 355, 344, 370, 337, 363, 352]

# 1. Conversion en GPa — 1 GPa = 1000 MPa
Re_gpa = [v / 1000 for v in Re]
print("1. En GPa : ", Re_gpa)

# 2. Valeurs conformes — limite nominale S355 est 355 MPa
Re_conforms = [v for v in Re if v >= 355]
print("2. Conformes : ", Re_conforms)

# 3. Taux de conformité
taux = len(Re_conforms) / len(Re) * 100
print("3. Taux de conformité : ", round(taux, 1), "%")  # round(x, n) arrondit x à n décimales

# 4. Écarts à la moyenne — la somme doit être proche de 0 par définition de la moyenne
moy = moyenne(Re)
Re_centre = [v - moy for v in Re]
print("4. Somme des écarts : ", round(sum(Re_centre), 6))  # ≈ 0
````

## Partie 3 — Lecture de fichiers

### Exercice 3.1 — Lire un CSV d'essai de traction

Le fichier `donnees_traction.csv` contient des résultats d'essai de traction sur un acier 316L.
Les deux premières lignes sont des commentaires (commencent par `#`), puis vient l'en-tête, puis les données :

```
# Essai de traction — acier 316L
# Colonnes : deformation (sans unite), contrainte (MPa)
deformation,contrainte
0.0000,0.0
0.0005,105.0
...
```

**Votre tâche :**

1. Ouvrez le fichier avec `with open(...)` et lisez-le ligne par ligne.
2. Ignorez les lignes de commentaires (`#`) et la ligne d'en-tête.
3. Stockez les déformations dans une liste `deformations` et les contraintes dans `contraintes`.
4. Affichez le nombre de points lus, la déformation maximale, et la contrainte maximale.

> ⚠️ N'oubliez pas `strip()` **avant** `split(",")` et de convertir avec `float()`.

````python
# Exercice 3.1 — solution

deformations = []  # liste des déformations lues
contraintes  = []  # liste des contraintes lues

with open("donnees_traction.csv", "r") as f:
    for ligne in f:
        ligne = ligne.strip()         # supprimer \n et espaces en bout de ligne
        if ligne.startswith("#") or ligne == "":
            continue                  # ignorer commentaires et lignes vides
        if not ligne[0].isdigit():
            continue                  # ignorer l'en-tête (commence par une lettre)
        morceaux = ligne.split(",")   # découper sur la virgule
        deformations.append(float(morceaux[0]))
        contraintes.append(float(morceaux[1]))

print("Nombre de points : ", len(deformations))
print("Déformation max  : ", round(max(deformations), 4))
print("Contrainte max   : ", round(max(contraintes), 1), "MPa")
````

### Exercice 3.2 — Statistiques sur les données lues

En utilisant vos fonctions `moyenne()` et `ecart_type()` de la Partie 1, calculez les statistiques sur les **contraintes** lues dans l'exercice précédent.

Répondez également dans la cellule Markdown suivante :
1. Quel est le **module de Young** apparent (pente de la zone élastique) ?  
   *(Utilisez les deux premiers points : $E \approx \Delta\sigma / \Delta\varepsilon$)*
2. À quelle déformation environ la courbe quitte-t-elle le domaine élastique ?

````python
# Exercice 3.2 — solution

# Statistiques sur les contraintes lues dans l'exercice 3.1
print("Moyenne     : ", round(moyenne(contraintes), 1), "MPa")
print("Écart-type  : ", round(ecart_type(contraintes), 1), "MPa")

# Estimation du module de Young à partir des deux premiers points
# E ≈ Δσ / Δε  (pente de la zone linéaire initiale)
delta_sigma = contraintes[1] - contraintes[0]    # MPa
delta_eps   = deformations[1] - deformations[0]  # sans unité
E_mpa = delta_sigma / delta_eps                  # MPa
E_gpa = E_mpa / 1000                             # GPa
print("Module de Young estimé : ", round(E_gpa, 0), "GPa")
````

**Réponses :**

1. **Module de Young estimé : 210 GPa** — calculé à partir des deux premiers points :
   $E \approx \Delta\sigma / \Delta\varepsilon = 105{,}0\,\text{MPa} / 0{,}0005 = 210\,000$ MPa = 210 GPa.
   C'est cohérent avec la valeur connue pour l'acier inoxydable 316L (environ 193–210 GPa).

2. **Fin du domaine élastique vers $\varepsilon \approx 0{,}002$** — au-delà, la courbe s'aplatit nettement : la contrainte passe de 418 MPa à $\varepsilon = 0{,}002$ à 425 MPa à $\varepsilon = 0{,}0025$,
   alors que la pente linéaire prédit ≈ 525 MPa pour $\varepsilon = 0{,}0025$.
